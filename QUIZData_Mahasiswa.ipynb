```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\User\DATASCIENCE\Data_Mahasiswa.csv"
df = pd.read_csv(file_path, delimiter=";")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mengubah kolom "Programming Skill" menjadi numerik
df["Programming Skill"] = pd.to_numeric(df["Programming Skill"], errors='coerce')

# Plot histogram
plt.figure(figsize=(8, 5))
sns.histplot(df["Programming Skill"].dropna(), bins=10, kde=True, color='blue')
plt.xlabel("Programming Skill")
plt.ylabel("Jumlah Mahasiswa")
plt.title("Distribusi Programming Skill Mahasiswa")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Memilih 15 data teratas
df_top15 = df.head(15)

# Mendeteksi kolom numerik
numeric_cols = df_top15.select_dtypes(include=['number']).columns

# Menyesuaikan ukuran plot
plt.figure(figsize=(15, 6))

# Membuat histogram untuk semua kolom numerik
df_top15[numeric_cols].hist(bins=10, figsize=(15, 6), layout=(2, len(numeric_cols)//2))

plt.suptitle("Histogram dari 15 Data Teratas")
plt.show()

# Membuat boxplot
plt.figure(figsize=(15, 6))
df_top15[numeric_cols].plot(kind='box', subplots=True, layout=(2, len(numeric_cols)//2), figsize=(15, 6))

plt.suptitle("Boxplot dari 15 Data Teratas")
plt.show()
```
