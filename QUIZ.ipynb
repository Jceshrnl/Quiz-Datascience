```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\User\DATASCIENCE\mahasiswa.csv"
df = pd.read_csv(file_path)
```

```python
df.head(10)
```

```python
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['IPK'], bins=5, kde=True, color='blue')
plt.xlabel("IPK Mahasiswa")
plt.ylabel("Frekuensi")
plt.title("Distribusi IPK Mahasiswa")
plt.show()
```

```python
# Mengelompokkan IPK ke dalam kategori
bins = [0, 2.5, 3.0, 3.5, 4.0]
labels = ["Rendah", "Sedang", "Tinggi", "Sangat Tinggi"]
df["Kategori_IPK"] = pd.cut(df["IPK"], bins=bins, labels=labels, include_lowest=True)

# Menghitung jumlah mahasiswa dalam setiap kategori
ipk_counts = df["Kategori_IPK"].value_counts().sort_index()

# Membuat Bar Chart
plt.figure(figsize=(8, 5))
sns.barplot(x=ipk_counts.index, y=ipk_counts.values, palette="Blues_r")
plt.xlabel("Kategori IPK")
plt.ylabel("Jumlah Mahasiswa")
plt.title("Jumlah Mahasiswa Berdasarkan Kategori IPK")
plt.show()
```

```python
np.random.seed(42)
df["Tahun_Masuk"] = np.random.randint(2018, 2023, size=len(df))

# Menghitung rata-rata IPK per tahun
ipk_per_tahun = df.groupby("Tahun_Masuk")["IPK"].mean()

# Membuat Line Graph
plt.figure(figsize=(8, 5))
sns.lineplot(x=ipk_per_tahun.index, y=ipk_per_tahun.values, marker='o', color='red')
plt.xlabel("Tahun Masuk")
plt.ylabel("Rata-rata IPK")
plt.title("Perkembangan Rata-rata IPK Mahasiswa per Tahun")
plt.show()
```

```python
# Membuat Pie Chart untuk proporsi kategori IPK
plt.figure(figsize=(7, 7))
plt.pie(ipk_counts, labels=ipk_counts.index, autopct='%1.1f%%', colors=['#66b3ff', '#99ff99', '#ffcc99', '#ff6666'])
plt.title("Proporsi Mahasiswa Berdasarkan Kategori IPK")
plt.show()
```

```python
# Menghitung ukuran pemusatan data
mean_ipk = df["IPK"].mean()  # Rata-rata
median_ipk = df["IPK"].median()  # Median
mode_ipk = df["IPK"].mode()[0]  # Modus
```
